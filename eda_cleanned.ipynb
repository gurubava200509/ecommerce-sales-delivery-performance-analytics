```python
import pandas as pd

df = pd.read_csv("C:\\Program Files\\MySQL\\MySQL Server 8.0\\Uploads\\olist_orders_dataset.csv")

print(df.head(10))
```

```python
df.info()
```

```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['order_approved_at'])
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df.to_csv("cleaned_orders.csv", index=False)
```

```python
status_counts = df['order_status'].value_counts()

print(status_counts)
```

```python
status_percent = df['order_status'].value_counts(normalize=True) * 100

print(status_percent)
```

```python
df['order_month'] = df['order_purchase_timestamp'].dt.to_period('M')
```

```python
monthly_orders = df.groupby('order_month').size()

print(monthly_orders)
```

```python
monthly_orders = df.groupby('order_month').size().sort_index()

print(monthly_orders)
```

```python
monthly_orders = df.groupby('order_month').size().reset_index(name='order_count')

print(monthly_orders)
```

```python
df['delivery_days'] = (
    df['order_delivered_customer_date'] - df['order_purchase_timestamp']
).dt.days
```

```python
print(df['delivery_days'].describe())
```

```python
delivery_df = df.dropna(subset=['delivery_days'])
```

```python
print(delivery_df['delivery_days'].describe())
```

```python
df['delivery_delay'] = (
    df['order_delivered_customer_date'] - df['order_estimated_delivery_date']
).dt.days
```

```python
print(df['delivery_delay'].describe())
```

```python
late_orders = (df['delivery_delay'] > 0).sum()
on_time_orders = (df['delivery_delay'] <= 0).sum()

print("Late Orders:", late_orders)
print("On-time / Early Orders:", on_time_orders)
```

```python
late_percent = (df['delivery_delay'] > 0).mean() * 100

print("Late Delivery %:", late_percent)
```

```python
df = df[df['delivery_days'] < 60]
```

```python
df['order_year'] = df['order_purchase_timestamp'].dt.year
df['order_month_num'] = df['order_purchase_timestamp'].dt.month
```
